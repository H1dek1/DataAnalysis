# Gini Impurity

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
delta_x = 0.001
p = np.arange(0, 1+delta_x, delta_x)
```

```python
gini = 1 - (p**2 + (1-p)**2)
```

```python
fig, ax = plt.subplots(1,1, figsize=(8, 6))
ax.set_xlabel(r'$n_A/N_{AB}$', fontsize=20)
ax.set_ylabel(r'$I_{Gini}}$', fontsize=20)
ax.plot(p, gini)
```
